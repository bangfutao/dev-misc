```python
# ======================================================================
# Solve Eight Queens on 8x8 chess board without attcking each other
# author: Bangfu Tao
# ======================================================================
import numpy as np
M = 8
matrix = np.zeros((M, M))
num_solutions = 0
```

```python
def clear_row(r):
    for c in range(M):
        matrix[r][c] = 0

def set_board(r, c):
    clear_row(r)
    matrix[r][c] = 1

def print_solution():
    print(f"\nsolution {num_solutions}:")
    for i in range(M):
        text = ''
        for j in range(M):
            if (matrix[i][j] == 0):
                text += "- "
            else:
                text += "Q "
        print(text)

def is_safe(row, col):
    for r in range(row):
        if (matrix[r][col] > 0):
            return False  # attack from same column

        if ((col + r - row) >= 0 and (col + r - row) < M
            and matrix[r][col + r - row] > 0):
            return False;  # attack from upper-left
        
        if ((col + row - r) >= 0 and (col + row - r) < M
            and matrix[r][col + row -r] > 0):
            return False;  # attack from upper-right
        
    return True  # square(row, col) is safe

def solve(row):
    global num_solutions
    if (row >= M):
        num_solutions += 1
        print_solution()
        return;
    for col in range(M):
        if (is_safe(row, col)):
            set_board(row, col)
            solve(row + 1)
```

```python
# Call the function to get solutions
solve(0)
```
